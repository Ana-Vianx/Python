## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
import locale 

locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8') #define a moeda como real brasileiro

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]
vendas = vendas_1sem + vendas_2sem #junta duas listas

maior = max(vendas) #valor máximo da lista
menor = min(vendas) #valor mínimo da lista

i_menor = vendas.index(menor) #índice do menor valor da lista
i_maior = vendas.index(maior) #índice do maior valor da lista

mes_maior = meses[i_maior] #pega o mes com maior faturamento usando o índice do maior valor
mes_menor = meses[i_menor] #pega o mes com menor faturamento usando o índice do menor valor

fat_anual = sum(vendas) #soma todos os valores da lista de venda do ano
porcnt = float(maior)/float(fat_anual) #porcentagem de contribuição do mes com maior faturamento no resultado anual

print('O maior faturamento do ano foi {} no mês de {}.'.format(locale.currency(maior, grouping=True), mes_maior)) #locale converte o float em real brasileiro
print('O menor faturamento do ano foi {} no mês de {}.'.format(locale.currency(menor, grouping=True), mes_menor))
print('Com o total anual fechando em {} e o mês de {} sendo {:.2%} desse total.'.format(locale.currency(fat_anual, grouping=True), mes_maior, porcnt))                         

```

## 3. Crie uma lista com o top 3 valores de vendas do ano (sem fazer "no olho")

Dica: o método remove retira um item da lista.

```python
import locale 

locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8') #define a moeda como real brasileiro

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]
vendas = vendas_1sem + vendas_2sem #junta duas listas

top3 = sorted(vendas, reverse=True)[:3] #sorted ordena do maior pro menor e o [:3] pega os 3 primeiros valores maiores
moeda = [locale.currency(v, grouping=True) for v in top3] #transforma os valores em moeda brasileira

print('Os 3 maiores valores de faturamento no ano foram: {}'.format(moeda))
```
